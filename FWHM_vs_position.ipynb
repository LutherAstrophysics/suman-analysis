```python
from trout.files.reference_log_file import ReferenceLogFile

import matplotlib.pyplot as plt
```

```python
f = ReferenceLogFile.get_ref_revised_71()
```

```python
star_ranges = {
    "1-100": range(1, 101),
    "100-300": range(100, 301),
    "300-600": range(300, 601),
    "600-1200": range(600, 1201),
    "1200-2508": range(1200, 2508)
    }
```

```python
def by_position(pos="X"):
    if pos not in ["X", "Y"]:
        raise Exception("Only positions are 'X' and 'Y'")
    pos_index = 0 if pos == "X" else 1
    x_label = "X-Coordinate" if pos == "X" else "Y-Coordinate"
    for k, v in star_ranges.items():
        x = []
        y = []
        for s in v:
            star_fwhm = f.get_star_fwhm(s)
            star_x = f.get_star_xy(s)[pos_index]
            if star_fwhm > 0:
                x.append(star_x)
                y.append(star_fwhm)

        plt.plot(x, y, "ro")
        plt.xlabel(x_label)
        plt.ylabel("FWHM")
        plt.title(f"Star: {k}")
        plt.show()
```

```python
by_position("X")
```

```python
by_position("Y")
```
