```python
# Find stars that present in df_2022 but not in current reference file
```

```python
def is_star_present_in_current_ref(x, y, threshold_distance=2, silent=False):
    found = False
    candidate = 1
    
    while not found and candidate < 2509:
        cx, cy = stars_positions_dict[candidate]
        distance = ((cx - x)**2 + (cy - y)**2)**0.5
        found = distance < threshold_distance
        if found and not silent:
            print("Star", candidate, cx, cy)
        candidate += 1
        
    return found
```

```python
def get_star_in_newly_generated_2003_df(old_star_no, threshold=1):
    x, y = stars_positions_dict[old_star_no]
    return df_2003.loc[
        (df_2003["xcentroid"] < x+threshold)
        & (df_2003["xcentroid"] > x-threshold)
        & (df_2003["ycentroid"] > y-threshold)
        & (df_2003["ycentroid"] < y+threshold)
    ]
```

```python
def get_star_in_new_ref(old_star_no, threshold=1):
    x, y = stars_positions_dict[old_star_no]
    return df_2022.loc[
        (df_2022["xcentroid"] < x+threshold)
        & (df_2022["xcentroid"] > x-threshold)
        & (df_2022["ycentroid"] > y-threshold)
        & (df_2022["ycentroid"] < y+threshold)
    ]
```

```python
new_x, new_y = df_2022["xcentroid"], df_2022["ycentroid"]
new_star_positions = zip(new_x, new_y)
new = []
found = 1
for index, star_new in enumerate(new_star_positions, start=1):
    if is_star_present_in_current_ref(*star_new, threshold_distance=4, silent=True):
        pass
    else:
        print(found, "new star", index, "position", star_new, "Y-flipped", star_new[0], f"{(1024-star_new[1]):.2f}")
        new.append((index, star_new))
        found += 1
print("total new stars", len(new))
```
