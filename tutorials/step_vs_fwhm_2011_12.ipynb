```python
from trout.files.reference_log_file import ReferenceLogFile
from trout.stars import get_star
import numpy as np

import matplotlib.pyplot as plt
```

```python
f = ReferenceLogFile.get_ref_revised_71()
```

```python
f.get_star_fwhm(1)
```

```python
s1 = get_star(1)
```

```python
s1.step(2011, 2012)
```

```python
all_stars = [i for i in range(1, 101)]
```

```python
x = [] # FWHM
y = [] # Step
```

```python
for star in all_stars:
    stp = get_star(star).step(2011, 2012)
    fwhm = f.get_star_fwhm(star)
    if not np.isnan(stp) and fwhm:
        x.append(fwhm)
        y.append(stp)
```

```python
plt.plot(x, y, 'ro')
plt.xlabel('FWHM')
plt.ylabel('Step')
plt.title(f'Stars 1-100 step 2011-12')
```
