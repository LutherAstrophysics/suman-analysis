```python
import matplotlib.pyplot as plt
import numpy as np

from trout.color import get_color
from trout.stars import get_star, Star
from trout.nights.year_nights import get_nights_in_a_year
```

```python
START_YEAR = 2003
END_YEAR = 2022
```

```python
VALID_YEARS = []
for y in range(START_YEAR, END_YEAR + 1):
    if len(get_nights_in_a_year(y)) > 0:
        VALID_YEARS.append(y)
```

```python
def star_attendance_plot(star_no, data_fn=Star.mean):
    s = get_star(star_no)
    attendances = np.array([s.attendance(y) for y in VALID_YEARS])
    signals = []
    for y in VALID_YEARS:
        s.select_year(y)
        signals.append(data_fn(s))
        
    fig, ax = plt.subplots()
    ax.scatter(VALID_YEARS, signals, s=attendances*100, alpha=0.5)
    
    
    ax.set_xlabel("Year", fontsize=15)
    ax.set_ylabel(f"{data_fn.__name__}", fontsize=15)
    ax.set_title(f"Star {star_no}")
    
    ax.grid(True)
    fig.tight_layout()
    
    plt.show()
```

```python
star_attendance_plot_by_mean = lambda star : star_attendance_plot(star, data_fn=Star.mean)
star_attendance_plot_by_median = lambda star : star_attendance_plot(star, data_fn=Star.median)
```
