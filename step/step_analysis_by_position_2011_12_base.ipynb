```python
from trout.vis import preview_step
from trout.files.reference_log_file import ReferenceLogFile
```

```python
f = ReferenceLogFile.get_ref_revised_71()
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
def ps(start_star, end_star):
    """
    Wrapper around preview_step
    """
    preview_step(start_star, 
                 end_star, 
                 2011, 
                 2012, 
                 get_x=lambda x : f.get_star_xy(x)[0], 
                 x_label="X-Position",
                 title=f"Stars: {start_star}-{end_star}"
                )
```

```python
star_ranges = {
    "1-100": range(1, 101),
    "100-300": range(100, 301),
    "300-600": range(300, 601),
    "600-1200": range(600, 1201),
    "1200-2508": range(1200, 2509),
    "1-2508": range(1, 2509) # See all data together
    }

for k, v in star_ranges.items():
    start, end = v[0], v[-1]
    ps(start, end)
```
