```python
FROM=2011
TO=2012

def save_table(start, end, file_to_save):
    headers = [
        "Star_no",
        "Old_step",
        "New_step",
        "Step_factor",
        "Color",
        "InternightBand",
        "X",
        "Y",
        "DupStar",
        "Neighbors"

    ]
    to_write = []
    to_write.append("".join(map(lambda x : f"{x:<25s}", headers)) + "\n")
    
    for star_no in range(start, end + 1):
        x, y = f.get_star_xy(star_no)
        star = Star(star_no)
        from_year_mean = star.select_year(FROM).mean()
        to_year_mean = star.select_year(TO).median()
        old_step = to_year_mean / from_year_mean
        step_factor = step_factors[star_no]
        new_step = to_year_mean / (from_year_mean * step_factor)
        internight_band = star.get_internight_band()
        color = star.get_color()
        neighbors = list(map(lambda x: x[0], star.closest_neighbors(filter_fn=lambda x: x[1] < 20)))
        neighbors_str = ",".join(map(str, neighbors))
        dup = star.brightest_duplicate()
        
        if dup:
            dup = dup[0]
            dup = f"{dup:<25.2f}"
        else:
             dup = pd.NA
            
        if color:
            color = f"{color:<25.2f}"
        else:
            color = pd.NA
            
        record = (
            f"{star_no:<25d}"
            f"{old_step:<25.4f}"
            f"{new_step:<25.4f}"
            f"{step_factor:<25.4f}"
            f"{color:<25s}"
            f"{internight_band:<25s}"
            f"{x:<25.2f}"
            f"{y:<25.2f}"
            f"{dup:<25s}"
            f"{neighbors_str:<25s}"
        )
        to_write.append(record+"\n")
    
    with open(file_to_save, 'w') as fd:
        fd.writelines(to_write)
```
